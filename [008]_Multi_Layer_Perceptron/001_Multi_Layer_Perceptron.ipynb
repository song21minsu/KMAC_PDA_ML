# Multi-Layer Perceptron

# I. Perceptron : AND, OR

* y = WX + b
* W와 X는 행렬

> ## 1) Activation Function - sigmoid( ) 

```
import numpy as np

def sigmoid(x):
    y = 1 / (1 + np.exp(-x))

    if y > 0.5:
      z = 1
    else:
      z = 0

    return z
```

> ## 2) AND

* y = sigmoid(WX + b)
* W = [5, 5], b = -6


```
# X = [0, 0] / False and False
# sigmoid(np.array([5, 5]) @ np.array([[0], [0]]) - 6)

sigmoid(np.dot(np.array([5, 5]), np.array([[0], [0]])) - 6)
```

```
# X = [0, 1] / False and True

sigmoid(np.dot(np.array([5, 5]), np.array([[0], [1]])) - 6)
```

```
# X = [1, 0] / True and False

sigmoid(np.dot(np.array([5, 5]), np.array([[1], [0]])) - 6)
```

```
# X = [1, 1] / True and True

sigmoid(np.dot(np.array([5, 5]), np.array([[1], [1]])) - 6)
```

> ## 3) OR

* y = sigmoid(WX + b)
* W = [6, 6], b = -3

```
# X = [0, 0] / False or False

sigmoid(np.dot(np.array([6, 6]), np.array([[0], [0]])) - 3)
```

```
# X = [0, 1] / False or True

sigmoid(np.dot(np.array([6, 6]), np.array([[0], [1]])) - 3)
```

```
# X = [1, 0] / True or False

sigmoid(np.dot(np.array([6, 6]), np.array([[1], [0]])) - 3)
```

```
# X = [1, 1] / True or True

sigmoid(np.dot(np.array([6, 6]), np.array([[1], [1]])) - 3)
```

# II. Multi-Layer Perceptron : XOR

* Activation : sigmoid()
* y1 = W1X + b1
* y2 = W2X + b2
* y = W3Y + b3
* W1, W2, W3, X, Y는 행렬
 * W1 = [3, 3], b1 = -5
 * W2 = [-8, -8], b2 = 4
 * W3 = [-9, -9], b3 = 9
 * Y = [y1, y2]


> ## 1) Input Layer : X = [0, 0] / False xor False

* Hidden Layer

```
y1 = sigmoid(np.array([3, 3]) @ np.array([[0], [0]]) - 5)
y2 = sigmoid(np.array([-8, -8]) @ np.array([[0], [0]]) + 4)

print(y1, y2)
```

* Output Layer

```
sigmoid(np.array([-9, -9]) @ np.array([[y1], [y2]]) + 9)
```

> ## 2) Input Layer :  X = [0, 1] / False xor True

* Hidden Layer

```
y1 = sigmoid(np.array([3, 3]) @ np.array([[0], [1]]) - 5)
y2 = sigmoid(np.array([-8, -8]) @ np.array([[0], [1]]) + 4)

print(y1, y2)
```

* Output Layer

```
sigmoid(np.array([-9, -9]) @ np.array([[y1], [y2]]) + 9)
```

> ## 3) Input Layer :  X = [1, 0] / True xor False

* Hidden Layer

```
y1 = sigmoid(np.array([3, 3]) @ np.array([[1], [0]]) - 5)
y2 = sigmoid(np.array([-8, -8]) @ np.array([[1], [0]]) + 4)

print(y1, y2)
```

* Output Layer

```
sigmoid(np.array([-9, -9]) @ np.array([[y1], [y2]]) + 9)
```

> ## 4) Input Layer :  X = [1, 1] / True xor True

* Hidden Layer

```
y1 = sigmoid(np.array([3, 3]) @ np.array([[1], [1]]) - 5)
y2 = sigmoid(np.array([-8, -8]) @ np.array([[1], [1]]) + 4)

print(y1, y2)
```

* Output Layer

```
sigmoid(np.array([-9, -9]) @ np.array([[y1], [y2]]) + 9)
```
